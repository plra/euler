```python
def targets(xs):
    if len(xs) == 1:
        return xs
    ts = set()
    for x in xs:
        other_xs = xs.copy()
        other_xs.remove(x)
        subtargets = targets(other_xs)
        for t in subtargets:
            new_ts = {x + t, x - t, x * t, t - x}
            if t != 0: new_ts.add(x / t)
            if x != 0: new_ts.add(t / x)
            ts |= new_ts
    return ts

print(targets({1}))
print(targets({1, 2, 3, 4}))
```

```python
def run_length(xs):
    i = 0
    while i + 1 in xs:
        i += 1
    return i
```

```python
digits, best_run_length = None, 0

for a in range(1, 10):
    for b in range(a+1, 10):
        for c in range(b+1, 10):
            for d in range(c+1, 10):
                ts = targets({a, b, c, d})
                if run_length(ts) >= best_run_length:
                    digits, best_run_length = {a, b, c, d}, run_length(ts)

digits, best_run_length
```
