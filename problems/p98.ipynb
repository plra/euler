```python
from math import sqrt, floor, ceil
from itertools import combinations
```

```python
words = []

with open("../res/words.txt") as f:
    words = [w.replace('"', "") for w in f.readline().split(",")]

# words.append("CARE")
# words.append("RACE")
words = sorted(words, key=lambda s: -len(s))
words
```

```python
pairs = []
for i, w1 in enumerate(words):
    for w2 in words[i + 1 :]:
        if sorted(w1) == sorted(w2):
            pairs.append((w1, w2))
pairs
```

```python
MAX_LEN = 9
anagrams = {}
for x in range(1, int(sqrt(10**MAX_LEN - 1))):
    d = len(str(x**2))
    if d not in anagrams:
        anagrams[d] = dict()
    ss = ''.join(sorted(str(x**2)))
    if ss not in anagrams[d]:
        anagrams[d][ss] = set()
    anagrams[d][ss].add(x**2)

# Purge self-anagrams
anagrams = {d: {ss: anas for ss, anas in anagrams[d].items() if len(anas) > 1} for d in anagrams}
```

```python
# Precondition: w1, w2 are anagrams
def sawps(w1, w2):
    d = len(str(w1))
    pairs = []
    for _, anas in anagrams[d].items():
        for n1, n2 in combinations(anas, 2):
            sub = {c: n for c, n in zip(str(w1), str(n1))}
            valid_sub = True
            for c, n in sub.items():
                for cp, np in sub.items():
                    if n == np and c != cp:
                        valid_sub = False
#             if valid_sub:
#                 print('Trying {}={}, {}={}'.format(w1, n1, w2, n2))
            if valid_sub and int(''.join([sub[c] for c in str(w2)])) == n2:
                pairs.append((n1, n2))
    return pairs

sawps('CARE', 'RACE')
```

```python
for w1, w2 in pairs:
    print('{}, {}: {}'.format(w1, w2, sawps(w1, w2)))
```
